# Functional data cleaning: The Why

## Learning Objectives

Our fundamental rules for data cleaning are:

1. Start with an empty DataFrame
1. Touch every variable just once
1. Touch with a pure function

After working through this topic, you should be able to explain how these rules help
achieve the following five goals:

- Explain the importance of never modifying source data
- Discuss the importance of separating data management and analysis
- Explain how functions can help to make data management code readable and testable
- Discuss strategies to minimize side effects in data management code
- Discuss strategies to make data management in notebooks more robust
